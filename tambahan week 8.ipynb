```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
import sympy as sp
import control as ct
from control.matlab import tf2ss, ss2zpk
from scipy.linalg import eig
```

```python
# Parameter dari rangkaian Sallen-Key
R = 1e3       # 1 kOhm
C = 0.022e-6  # 0.022 uF

# Frekuensi natural
wn = 1 / (R * C)  # rad/s

# Variasi gain K
K_values = (1.05, 1.1, 1.15, 1.2, 1.25, 1.3, 1.35, 1.4, 1.45, 1.5, 1.55, 1.6, 1.65, 1.7, 1.75)

plt.figure()
for i, (K) in enumerate(K_values):
    zeta = (3 - K) / (2 * np.sqrt(K))  # Rasio redaman khas Sallen-Key
    num = [K * wn**2]
    den = [1, 2*zeta*wn, wn**2]
    G = ct.TransferFunction(num, den)
    cplt = ct.root_locus_plot(
        G,
        grid = True,
        title = "Root Locus of G(s)",
    )

ax = cplt.axes[0, 0]
ax.set_xlabel('Real axis')
ax.set_ylabel('Imaginary axis')


ax.legend()
plt.show()
```

```python
# Parameter dari rangkaian Sallen-Key
R = 1e3       # 1 kOhm
C = 0.022e-6  # 0.022 uF

# Frekuensi natural
wn = 1 / (R * C)  # rad/s

# Variasi gain K
K_values = (1.05, 1.1, 1.15, 1.2, 1.25, 1.3, 1.35, 1.4, 1.45, 1.5, 1.55, 1.6, 1.65, 1.7, 1.75) 

# Step response gabungan
plt.figure(figsize=(10, 6))
for K in K_values:
    zeta = (3 - K) / (2 * np.sqrt(K))  # Rasio redaman khas Sallen-Key
    num = [K * wn**2]
    den = [1, 2*zeta*wn, wn**2]
    G = ct.TransferFunction(num, den)
    t, y = ct.step_response(G)
    plt.plot(t, np.squeeze(y), label=f"K = {K}")
plt.title("Step Response Filter Sallen-Key untuk Berbagai Gain K")
plt.xlabel("Waktu [s]")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Step response detail per K
for K in K_values:
    zeta = (3 - K) / (2 * np.sqrt(K))
    num = [K * wn**2]
    den = [1, 2*zeta*wn, wn**2]
    G = ct.TransferFunction(num, den)
    t, y = ct.step_response(G)
    y = np.squeeze(y)
    info = ct.step_info(G)

    rise_time = info.get("RiseTime", np.nan)
    settling_time = info.get("SettlingTime", np.nan)
    overshoot = info.get("Overshoot", np.nan)
    peak_value = info.get("Peak", np.nan)

    plt.figure(figsize=(10, 5))
    plt.plot(t, y, label="Respons Step")
    if not np.isnan(rise_time):
        plt.axvline(rise_time, color='g', linestyle='--', label=f"Rise Time: {rise_time:.4f}s")
    if not np.isnan(settling_time):
        plt.axvline(settling_time, color='orange', linestyle='--', label=f"Settling Time: {settling_time:.4f}s")
    if not np.isnan(peak_value):
        plt.axhline(peak_value, color='r', linestyle='--', label=f"Overshoot: {overshoot:.2f}%" if not np.isnan(overshoot) else "Peak")

    plt.title(f"Step Response - Gain K = {K}")
    plt.xlabel("Waktu [s]")
    plt.ylabel("Amplitudo")
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()

    print(f"\n=== Info Step Response untuk K = {K} ===")
    print ("zeta: ", zeta)
    for key, val in info.items():
        print(f"{key:15}: {val:.4f}")
```
