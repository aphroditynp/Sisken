```python
# Gain tetap
K = 1  

# Daftar variasi R dan C
variations = [
    (1e3, 1e-6),   # R = 1kΩ, C = 1µF
    (2e3, 1e-6),
    (3e3, 1e-6),
    (4e3, 1e-6),
    (5e3, 1e-6),
    (1e3, 1e-6),
    (1e3, 2e-6),
    (1e3, 3e-6),
    (1e3, 4e-6),
    (1e3, 5e-6),
]

plt.figure(figsize=(10, 6))

for R, C in variations:
    a = 1 / (R * C)
    G = ct.TransferFunction([K], [1, a])
    t, y = ct.step_response(G)
    y = np.squeeze(y)
    label = f"R={R/1e3:.0f}kΩ, C={C*1e6:.0f}µF"
    plt.plot(t, y, label=label)

# Set detail grafik
plt.title("Step Response of System with Variasi R dan C (K Tetap)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure()
for i, (R, C) in enumerate(variations):
    a = 1 / (R * C)
    G = ct.TransferFunction([K], [1, a])
    t, y = ct.step_response(G)
    # Step Response Information
    data = ct.step_info(G)

    # Ambil parameter penting
    rise_time = data["RiseTime"]
    settling_time = data["SettlingTime"]
    overshoot = data["Overshoot"]
    peak_time = data["PeakTime"]
    peak_value = data["Peak"]
    y = np.squeeze(y)
    plt.figure(figsize=(10, 5))
    plt.plot(t, y.T)
    # Tambahkan garis Rise Time
    plt.axvline(rise_time, color='g', linestyle='--', label=f"Rise Time: {rise_time:.4f}s")

# Tambahkan garis Settling Time
    plt.axvline(settling_time, color='orange', linestyle='--', label=f"Settling Time: {settling_time:.4f}s")

# Tambahkan garis Overshoot
    plt.axhline(peak_value, color='r', linestyle='--', label=f"Overshoot: {overshoot:.2f}%")
    plt.title("Step Response of the system after changing R and C")
    plt.xlabel("Time [s]")
    plt.ylabel("Apmplitude")
    plt.grid()
    plt.show()
    print("\n=== Step Response Information ===")
    for k, v in data.items():
        print(f'{k} : {float(v):.4f}')
```

```python
# Parameter dari rangkaian Sallen-Key
R = 1e3       # 1 kOhm
C = 0.022e-6  # 0.022 uF

# Frekuensi natural
wn = 1 / (R * C)  # rad/s

# Variasi gain K
K_values = range(2, 17)  # K dari 2 sampai 16

# Step response gabungan
plt.figure(figsize=(10, 6))
for K in K_values:
    zeta = (3 - K) / (2 * np.sqrt(K))  # Rasio redaman khas Sallen-Key
    num = [K * wn**2]
    den = [1, 2*zeta*wn, wn**2]
    G = ct.TransferFunction(num, den)
    t, y = ct.step_response(G)
    plt.plot(t, np.squeeze(y), label=f"K = {K}")
plt.title("Step Response Filter Sallen-Key untuk Berbagai Gain K")
plt.xlabel("Waktu [s]")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Step response detail per K
for K in K_values:
    zeta = (3 - K) / (2 * np.sqrt(K))
    num = [K * wn**2]
    den = [1, 2*zeta*wn, wn**2]
    G = ct.TransferFunction(num, den)
    t, y = ct.step_response(G)
    y = np.squeeze(y)
    info = ct.step_info(G)

    rise_time = info.get("RiseTime", np.nan)
    settling_time = info.get("SettlingTime", np.nan)
    overshoot = info.get("Overshoot", np.nan)
    peak_value = info.get("Peak", np.nan)

    plt.figure(figsize=(10, 5))
    plt.plot(t, y, label="Respons Step")
    if not np.isnan(rise_time):
        plt.axvline(rise_time, color='g', linestyle='--', label=f"Rise Time: {rise_time:.4f}s")
    if not np.isnan(settling_time):
        plt.axvline(settling_time, color='orange', linestyle='--', label=f"Settling Time: {settling_time:.4f}s")
    if not np.isnan(peak_value):
        plt.axhline(peak_value, color='r', linestyle='--', label=f"Overshoot: {overshoot:.2f}%" if not np.isnan(overshoot) else "Peak")

    plt.title(f"Step Response - Gain K = {K}")
    plt.xlabel("Waktu [s]")
    plt.ylabel("Amplitudo")
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()

    print(f"\n=== Info Step Response untuk K = {K} ===")
    print ("zeta: ", zeta)
    for key, val in info.items():
        print(f"{key:15}: {val:.4f}")
```
